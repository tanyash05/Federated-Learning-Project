CDP = better utility,

LDP = more privacy, but more noise/accuracy loss.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

original_data = np.random.normal(loc=0, scale=1.0, size=10000) # loc=mean; scale=std_dev; size of sample=10000
central_dp_data = original_data + np.random.normal(loc=0, scale=0.5, size=10000) # smaller noise = lower privacy, higher utility(accuracy)
local_dp_data = original_data + np.random.normal(loc=0, scale=1.5, size=10000)
```

```python
plt.hist(original_data, bins=100, alpha=0.5, label="Original", color='gray', density=True)
plt.hist(central_dp_data, bins=100, alpha=0.5, label="Central DP", color='blue', density=True)
plt.hist(local_dp_data, bins=100, alpha=0.5, label="Local DP", color='green', density=True)
plt.title("Noise Distribution: Central vs Local DP")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.show()
```

```python
x = np.linspace(-6,6,1000)
plt.plot(x, norm.pdf(x, np.mean(original_data), np.std(original_data)), label="Original", color='gray', linewidth=2)
plt.plot(x, norm.pdf(x, np.mean(central_dp_data), np.std(central_dp_data)), label="Central DP", color='blue', linewidth=2)
plt.plot(x, norm.pdf(x, np.mean(local_dp_data), np.std(local_dp_data)), label="Local DP", color='green', linewidth=2)
plt.title("Gaussian Curve Comparison")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.grid(True)
plt.show()
```

Curve -> (Narrow/Tall) -> Lower Privacy -> More accuracy

Curve -> (Wider) -> Strong Privacy -> Less Accuracy

```python
sigmas=[0.1,0.5,1.0,1.5]
x = np.linspace(-6,6,1000)

for sigma in sigmas:
    plt.plot(x, norm.pdf(x, 0, sigma), label=f'σ={sigma}')
plt.title("Effect of Sigma on Gaussian Noise")
plt.xlabel("Value")
plt.ylabel("Probability Density")
plt.legend()
plt.grid(True)
plt.show()
```
